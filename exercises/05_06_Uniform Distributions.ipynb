```python
import numpy as np

def generate_test_scores(num_scores, mean=70, std_dev=15, min_score=0, max_score=100):
    """
    Generate a list of normally distributed test scores.

    Parameters:
    - num_scores: The number of test scores to generate.
    - mean: The mean of the normal distribution (default is 70).
    - std_dev: The standard deviation of the normal distribution (default is 15).
    - min_score: The minimum possible score (default is 0).
    - max_score: The maximum possible score (default is 100).

    Returns:
    - A list of test scores.
    """
    # Generate normally distributed scores
    scores = np.random.normal(loc=mean, scale=std_dev, size=num_scores)

    # Clip the scores to be within the specified range
    scores = np.clip(scores, min_score, max_score)

    # Convert the numpy array to a list
    scores_list = scores.tolist()

    return scores_list

# Example usage
if __name__ == "__main__":
    num_scores = 100
    test_scores = generate_test_scores(num_scores)
    print(test_scores)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_test_scores(num_scores, mean=70, std_dev=15, min_score=0, max_score=100):
    """
    Generate a list of normally distributed test scores as integers.

    Parameters:
    - num_scores: The number of test scores to generate.
    - mean: The mean of the normal distribution (default is 70).
    - std_dev: The standard deviation of the normal distribution (default is 15).
    - min_score: The minimum possible score (default is 0).
    - max_score: The maximum possible score (default is 100).

    Returns:
    - A list of integer test scores.
    """
    # Generate normally distributed scores
    scores = np.random.normal(loc=mean, scale=std_dev, size=num_scores)

    # Clip the scores to be within the specified range
    scores = np.clip(scores, min_score, max_score)

    # Convert the scores to integers
    scores = scores.astype(int)

    # Convert the numpy array to a list
    scores_list = scores.tolist()

    return scores_list

def plot_scores_distribution(scores):
    """
    Plot a histogram of the test scores distribution.

    Parameters:
    - scores: A list of test scores.
    """
    plt.figure(figsize=(10, 6))
    plt.hist(scores, bins=range(0, 101, 5), edgecolor='black', alpha=0.7)
    plt.title('Distribution of Test Scores')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.xticks(range(0, 101, 5))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example usage
if __name__ == "__main__":
    num_scores = 100
    test_scores = generate_test_scores(num_scores)
    plot_scores_distribution(test_scores)
```
